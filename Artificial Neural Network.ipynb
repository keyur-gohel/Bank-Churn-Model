# **Bank Churn Modeling**


> Dataset consist data of bank's customers who has left the bank in last six months. We are building a classifier using **Artificial Neural Network (ANN)** to predict the customer will left bank or not based on given feature.



# Artificial Neural Network

### Importing Libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

### Get Dataset

```python
dataset = pd.read_csv('Churn_Modelling.csv')
x = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

## Encode Categorical Data

### Gender column

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

label_encoder = LabelEncoder()
x[:, 2] = label_encoder.fit_transform(x[:, 2])
```

### Country column

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

### Spliting the Dataset Into Training and Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## Create NN

### Initializing ANN

```python
ann = tf.keras.models.Sequential()
```

### Input Layer and First Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

### Adding Second Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

### Adding Output Layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

### Compiling the ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Fitting the ANN to Training Set

```python
ann.fit(x_train, y_train, batch_size = 10, epochs = 100)
```

### Predicting values

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
```

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)
```

### Confusion Matrix and Accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
